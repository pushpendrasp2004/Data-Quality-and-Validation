# Data Quality and Validation


1. Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

- Data Quality in ETL refers to the accuracy, completeness, consistency, reliability, and timeliness of data as it moves through extraction, transformation, and loading processes.

It is more than just data cleaning because:

Cleaning only removes errors (nulls, duplicates).
Data quality ensures:
Correct structure (schema validation)
Business rule compliance
Consistency across systems
Valid relationships (referential integrity)

Thus, data quality is a continuous process, not a one-time cleanup.

2.  Explain why poor data quality leads to misleading dashboards and incorrect decisions.

- Poor data quality results in:

Incorrect aggregations (wrong totals, averages)
Duplicate records inflating metrics
Missing values causing biased insights
Inconsistent formats leading to wrong grouping

Example:
If sales data has duplicates → dashboard shows inflated revenue → wrong business decisions.

“Garbage in = Garbage out”

3. What is duplicate data? Explain three causes in ETL pipelines

- Duplicate data = Same record appearing multiple times.

Causes:
Multiple data sources
Same record from different systems
Improper joins
Cartesian joins creating repeated rows
Reprocessing without checks
ETL job re-runs without deduplication logic

4. Differentiate between exact, partial, and fuzzy duplicates.
-

 | Type    | Description                         | Example                             |
| ------- | ----------------------------------- | ----------------------------------- |
| Exact   | Completely identical records        | Same Customer_ID, Product_ID, etc.  |
| Partial | Some fields match                   | Same customer but different address |
| Fuzzy   | Slight variations (spelling errors) | “Rahul Mehta” vs “Rahul Mehata”     |


5. Why should data validation be performed during transformation rather than after loading?

- Validation during transformation:

Prevents bad data from entering warehouse
Saves reprocessing cost
Enables real-time correction
Maintains pipeline reliability

Post-loading validation is too late and expensive

7. Write an SQL query on Sales_Transactions to list all duplicate keys and their counts using the
business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount )

- SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS duplicate_count
FROM Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING COUNT(*) > 1;
